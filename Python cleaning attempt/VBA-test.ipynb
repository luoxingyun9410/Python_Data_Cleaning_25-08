```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "/VBA 14_08_2023_ StockSoldReport.xls")
dataset
```

```python
import numpy as np
```

```python
dataset.nunique()
```

```python
columns_to_drop = dataset.nunique()[dataset.nunique()==0].index
columns_to_drop
```

```python
data_dropped_cols = dataset.copy()
data_dropped_cols = data_dropped_cols.drop(columns=columns_to_drop)
data_dropped_cols
```

```python
data_dropped_cols = data_dropped_cols.dropna(axis=1, how='all')
data_dropped_cols
```

```python
total_cols = data_dropped_cols.shape[1]
nan_percentage_per_row = data_dropped_cols.isna().sum(axis=1) / total_cols * 100

nan_percentage_per_row
```

```python
rows_to_keep = nan_percentage_per_row <= 92
filtered_df = data_dropped_cols[rows_to_keep]
filtered_df
```
