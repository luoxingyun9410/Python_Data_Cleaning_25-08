```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "/21_08_2023_ SalesmanCommissionReport.xls")
dataset
```

```python
def clean_excel_file(file_path):
    # Load the Excel file
    df = pd.read_excel(file_path)

    # Find the index of the row that contains the word "Description"
    description_index = df[df.applymap(lambda x: "Description" in str(x)).any(axis=1)].index.min()

    # Drop all rows above the description_index
    df_cleaned = df.iloc[description_index:]

    # Reset the index of the cleaned dataframe
    df_cleaned = df_cleaned.reset_index(drop=True)

    # Save the cleaned dataframe to a new Excel file with .xlsx format
    output_path = file_path.replace(".xls", "_cleaned.xlsx")
    df_cleaned.to_excel(output_path, index=False)

    return output_path
```

```python
# Use the function
output_file = clean_excel_file("H:\\Parrvans document\\Python cleaning attempt\\21_08_2023_ SalesmanCommissionReport.xls")
print(f"Cleaned file saved at: {output_file}")
```

```python
dataset_first_cleaned = pd.read_excel(pwd + "/21_08_2023_ SalesmanCommissionReport_cleaned.xlsx")
dataset_first_cleaned
```

```python
dataset_first_cleaned.columns = dataset_first_cleaned.iloc[0]
dataset_first_cleaned = dataset_first_cleaned [1:]
dataset_first_cleaned
```

```python
# Combining A and B columns to create the "New/Used" column
dataset_first_cleaned['New/Used'] = dataset_first_cleaned.iloc[:, 0] + " " + dataset_first_cleaned.iloc[:, 1]
dataset_first_cleaned
```

```python
# Identify the rows with "New Stock" and "Used Stock"
new_stock_row = dataset_first_cleaned[dataset_first_cleaned['New/Used'].str.contains('New Stock', na=False)].index[0]
used_stock_row = dataset_first_cleaned[dataset_first_cleaned['New/Used'].str.contains('Used Stock', na=False)].index[0]

# Create a new column 'Stock Type' and set its values based on the identified rows
dataset_first_cleaned['Stock Type'] = None
dataset_first_cleaned.loc[:used_stock_row, 'Stock Type'] = 'New'
dataset_first_cleaned.loc[used_stock_row:, 'Stock Type'] = 'Used'

# Remove rows that have "New Stock" and "Used Stock" as these are now redundant
dataset_first_cleaned = dataset_first_cleaned[~dataset_first_cleaned['New/Used'].str.contains('New Stock|Used Stock', na=False)]
dataset_first_cleaned
```

```python
dataset_first_cleaned = dataset_first_cleaned.dropna(axis=1, how='all')
dataset_first_cleaned
```

```python
# Removing all rows that contain the keyword "Total"
dataset_first_cleaned = dataset_first_cleaned[~dataset_first_cleaned.applymap(lambda x: "Total" in str(x)).any(axis=1)]
dataset_first_cleaned
```

```python
# Specify the path where you want to save the cleaned Excel file
output_path = "H:\\Parrvans document\\Python cleaning attempt\\First_Attempt_Python_SalesmanCommissionReport.xlsx"

# Export the DataFrame to an Excel file
dataset_first_cleaned.to_excel(output_path, index=False)

print(f"Cleaned dataset saved to: {output_path}")
```
